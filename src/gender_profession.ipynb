After analysing dataset with gender comparisons, its decided to select three professions: Surgeon, Nurse and Physician, as they all belong to  "medical" profile. 


```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Shilpi\BFH\MSE_Courses\Semester4\P2_BiasMitigation\New_again\P2_Code_New\csv_DS\bias_profession_gender.csv")
df.head()
```

```python
selected_professions = ['surgeon', 'nurse', 'physician']
```

```python
# Create separate datasets for each profession
df_surgeon = df[df['profession'] == selected_professions[0]]
df_nurse = df[df['profession'] == selected_professions[1]]
df_physician = df[df['profession'] == selected_professions[2]]
```

```python
df_physician.head()
df_nurse.head()
df_surgeon.head()
```

```python
print("Surgeon Dataset Size:", len(df_surgeon))
print("Nurse Dataset Size:", len(df_nurse))
print("Physician Dataset Size:", len(df_physician))
```

```python
# Save them as new CSV files
df_physician.to_csv("phycisian_ds.csv", index=False)
df_nurse.to_csv("nurse_ds.csv", index=False)
df_surgeon.to_csv("surgeon_ds.csv", index=False)
```

```python
# Count genders
gender_counts_prof1 = df_physician['gender'].value_counts()
gender_counts_prof2 = df_nurse['gender'].value_counts()
gender_counts_prof3 = df_surgeon['gender'].value_counts()

# Combine into one DataFrame for comparison
gender_comparison = pd.DataFrame({
    "physician": gender_counts_prof1,
    "nurse": gender_counts_prof2,
    "surgeon": gender_counts_prof3
}).fillna(0)  # fill missing genders with 0
```

```python
print(gender_comparison)
```

```python
import matplotlib.pyplot as plt

gender_comparison.plot(kind='bar', figsize=(8,5))
plt.ylabel("Count")
plt.title("Gender Comparison Across Professions")
plt.show()
```
