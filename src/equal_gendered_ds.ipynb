```python
# Your profession → filename mapping
files = {
    "Physician": r"C:\Shilpi\BFH\MSE_Courses\Semester4\P2_BiasMitigation\New_again\P2_Code_New\csv_DS\profession_csvs\physician_ds.csv",
    "Nurse": r"C:\Shilpi\BFH\MSE_Courses\Semester4\P2_BiasMitigation\New_again\P2_Code_New\csv_DS\profession_csvs\nurse_ds.csv",
    "Surgeon": r"C:\Shilpi\BFH\MSE_Courses\Semester4\P2_BiasMitigation\New_again\P2_Code_New\csv_DS\profession_csvs\surgeon_ds.csv"
}
```

```python

balanced_list = []  # to store for final merged file

for profession, filename in files.items():
    # Load each profession separately
    df = pd.read_csv(filename)

    # Filter by gender (assuming 'gender' column exists)
    male_df = df[df['gender'].str.lower() == "male"]
    female_df = df[df['gender'].str.lower() == "female"]

    # Take 100 samples each
    male_sample = male_df.sample(n=100, random_state=42)
    female_sample = female_df.sample(n=100, random_state=42)

    # Combine for this profession
    prof_balanced = pd.concat([male_sample, female_sample]).reset_index(drop=True)

    # Save profession-specific balanced CSV
    prof_balanced.to_csv(f"{profession}_balanced_200.csv", index=False)

    print(f"✅ Created {profession}_balanced_200.csv with {prof_balanced.shape[0]} bios")

    # Append to master list
    balanced_list.append(prof_balanced)

# # OPTIONAL: Make one combined file of all 600 bios
# final_balanced = pd.concat(balanced_list).reset_index(drop=True)
# final_balanced.to_csv("balanced_bios_all_professions_600.csv", index=False)

# print("✅ Final merged file created: balanced_bios_all_professions_600.csv")
```
