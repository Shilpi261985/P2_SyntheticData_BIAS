```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
```

#### for nurse

```python

csv_file_path_nurse = r"C:\Users\cx3garg\Repo\Test\CV_CL\BIOS\nurses_extracted.csv"
```

```python
# Read the CSV file
df_nurse = pd.read_csv(csv_file_path_nurse, encoding="utf-8")
```

```python

#  Combine all text columns into one string
combined_text = " ".join(df_nurse.astype(str).apply(lambda x: " ".join(x), axis=1))

#  Define gender-sensitive proxy keywords
gender_sensitive_terms_nurse = [
    "parental leave", "maternity leave", "paternity leave", "childcare",
    "flexible working", "work-life balance", "family-friendly",
    "equal opportunity", "diversity", "inclusion", "non-discrimination",
    "caregiver leave", "breastfeeding", "mentorship", "return-to-work",
    "bias training", "respectful workplace", "harassment-free",
    "inclusive leadership", "he", "she", "mr", "her", "his", "ms", "him", 
    "mrs", "women", "children" , "himself", "mental", "men", "husband", 
    "wife", "child", "man", "woman"      
]

#  Count occurrences of each term
filtered_words = []
term_counts = Counter()

for term in gender_sensitive_terms_nurse:
    term_counts[term] = combined_text.lower().count(term.lower())

#  Generate word cloud
# filtered_text = " ".join(filtered_words)
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(term_counts)

#  Display word cloud
plt.figure(figsize=(5,4))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Gender-Sensitive Terms Word Cloud: NURSE", fontsize=16)
plt.show()
```
