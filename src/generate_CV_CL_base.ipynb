```python
# import requests
import os
# from huggingface_hub import InferenceClient
# from huggingface_hub import login, hf_hub_download
from dotenv import load_dotenv
import openai
import pandas as pd
# from transformers import pipeline

```

```python
load_dotenv(".env.job")

# Load credentials
api_key = os.environ.get("AZURE_CHTDP_SWED_ENDPOINT_API_KEY")
endpoint = os.environ.get("AZURE_CHTDP_SWED_ENDPOINT")
deployment = os.environ.get("AZURE_CHAT_DEPLOYMENT_NAME")
```

```python
openai.api_type = "azure"
openai.api_key = api_key
openai.api_base = endpoint
openai.api_version = "2023-10-01-preview"
```

```python
def generate_text(prompt, max_tokens=500, temperature=0.7):
    response = openai.ChatCompletion.create(
        engine=deployment,  # Azure deployment name
        
        messages=[
            {"role": "system", "content": "You are an assistant who generates CVs and cover letters."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=max_tokens,
        temperature=temperature,
        n=1,
        stop=None
    )
    return response.choices[0].message.content.strip()
```

```python
# Load CSV with bios
bios_df = pd.read_csv(r"C:\Users\cx3garg\Repo\Test\CV_CL\BIOS\nurses_extracted.csv")  # replace with your CSV file path
job_ad = open(r"C:\Users\cx3garg\Repo\Test\CV_CL\job_ADs\nurse_job_ad.txt", encoding="utf-8").read()
```

```python
# Generate CV + cover letter for each bio
results = []

for idx, row in bios_df.iterrows():
    bio = row["hard_text"]
    gender = row["gender"]
    final_prompt = f"""Generate a CV and cover letter based on this bio:\n{bio}\nAnd this job ad:\n{job_ad}.
                    DO NOT return any AI-generated suggestions and notifications."""
    
    result_text = generate_text(final_prompt)
    results.append({
        "bio_index": idx,
        "cv_cover_letter": result_text,
        "gender": gender
    })
    print(f"✅ Generated CV + CL for bio {idx}")
```

```python
results_df = pd.DataFrame(results)
results_df.to_csv("generated_cv_cl_nurse.csv", index=False)
print("All CVs and cover letters saved to 'generated_cv_cl_nurse.csv'.")
```
